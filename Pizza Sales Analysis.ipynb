### Daily Treds- Total Orders

```python
ds['order_date']=pd.to_datetime(ds['order_date'],dayfirst=True)
ds['day_name']=ds['order_date'].dt.day_name()
weekday_order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
ds['day_name']=pd.Categorical(ds['day_name'],categories=weekday_order,ordered=True)
order_by_day=ds.groupby('day_name',observed=False)['order_id'].nunique()
ax=order_by_day.plot(kind='bar',figsize=(9,5),color='green', edgecolor='black')

plt.title('Total Orders per Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)

for i, var in enumerate(order_by_day):
    plt.text(i,var+20,str(var),ha='center',va='bottom',fontsize=9,fontweight='bold')

plt.tight_layout()
plt.show()
                    
```

### Daily Treds- Total Revenue

```python
ds['order_date']=pd.to_datetime(ds['order_date'],dayfirst=True)
ds['day_name']=ds['order_date'].dt.day_name()
weekday_order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
ds['day_name']=pd.Categorical(ds['day_name'],categories=weekday_order,ordered=True)
order_by_day=ds.groupby('day_name',observed=False)['total_price'].sum()
ax=order_by_day.plot(kind='bar',figsize=(9,5),color='blue', edgecolor='black')

plt.title('Total Revenue per Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)

for i, var in enumerate(order_by_day):
    plt.text(i,var+20,str(var),ha='center',va='bottom',fontsize=9,fontweight='bold')

plt.tight_layout()
plt.show()
```

### Daily Treds- Total Quantity


```python
ds['order_date']=pd.to_datetime(ds['order_date'],dayfirst=True)
ds['day_name']=ds['order_date'].dt.day_name()
weekday_order=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
ds['day_name']=pd.Categorical(ds['day_name'],categories=weekday_order,ordered=True)
order_by_day=ds.groupby('day_name',observed=False)['quantity'].sum()
ax=order_by_day.plot(kind='bar',figsize=(9,5),color='purple', edgecolor='black')

plt.title('Total Quantity per Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)

for i, var in enumerate(order_by_day):
    plt.text(i,var+20,str(var),ha='center',va='bottom',fontsize=9,fontweight='bold')

plt.tight_layout()
plt.show()
```

### Hourly Treds- Total Orders

```python
ds['order_time']=pd.to_datetime(ds['order_time'],format='%H:%M:%S')
ds['order_hour']=ds['order_time'].dt.hour

order_by_hour=ds.groupby('order_hour',observed=False)['order_id'].nunique()
ax=order_by_hour.plot(kind='bar',figsize=(9,5),color='purple', edgecolor='black')

plt.title('Total Order per Hour')
plt.xlabel('Hour of Day (24 Hours)')
plt.ylabel('Total Orders')
plt.xticks(rotation=0)

for i, var in enumerate(order_by_hour):
    plt.text(i,var+5,str(var),ha='center',va='bottom',fontsize=9,fontweight='bold')

plt.tight_layout()
plt.show()
```

### Monthly Treds- Total Orders

```python
ds['order_date']=pd.to_datetime(ds['order_date'],dayfirst=True)
ds['month_name']=ds['order_date'].dt.month_name()
month_order=["January","February","March","April","May","June","July","August","September","October","November","December"]
ds['month_name']=pd.Categorical(ds['month_name'],categories=month_order,ordered=True)
order_by_month=ds.groupby('month_name',observed=False)['order_id'].nunique()


plt.figure(figsize=(10,5))
plt.fill_between(order_by_month.index,order_by_month.values,color='Orange',alpha=0.6)
plt.plot(order_by_month.index,order_by_month.values,color='black',linewidth=2,marker='o')


plt.title('Total Orders per Month')
plt.xlabel('Month of the Year')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)

for i, var in enumerate(order_by_month):
    plt.text(i,var+20,str(var),ha='center',va='bottom',fontsize=9,fontweight='bold')

plt.tight_layout()
plt.show()
```

### % of Sales by Category

```python
category_sales = ds.groupby('pizza_category')['total_price'].sum()

category_pct = category_sales / category_sales.sum() * 100

plt.figure(figsize=(7,7))

colors = plt.get_cmap('tab20').colors

plt.pie(
    category_pct,
    labels=category_pct.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'width': 0.4}
)

plt.title('Percentage of Sales by Category')
plt.show()
```

### % of Sales by Pizza Size & Category

```python
sales_pivot = ds.pivot_table(
    index='pizza_category',
    columns='pizza_size',
    values='total_price',
    aggfunc='sum',
    fill_value=0
)

sales_pcl = sales_pivot / sales_pivot.sum().sum() * 100

plt.figure(figsize=(10,7))

sns.heatmap(
    sales_pcl,
    annot=True,
    fmt=".1f",
    cmap="YlOrRd",
    linewidths=0.5
)

plt.title('% of Sales by Pizza Size & Category')
plt.xlabel('Pizza Size')
plt.ylabel('Pizza Category')
plt.show()
```

### Total Pizza Sold by Pizza Category
